#### Naive approach

- template matching
        1. set of templates is stored in the system
        2. compare each template with every location on the input image 
        3. compute score for each location one the image
        4. score can be computed as cross correlation,sum of squared differences 

##### Remark
Template matching approach is sensitive to: illumination,perspective, deformation

#### Main problem with linear classification
   - accuracy drops if the classes are not separable using a hyperplane. 

#### Solution
   - the data can be transformed to a new space where classes in this new space are linearly separable,using one of the approaches below:
        1. an expert designs a function manually
        2. feature transformation function is learned from data (why not fully connected neural network ?)       - 

#### New approach:
 - convolutional networks learn a composite feature transformation function that makes classes of objects linearly separable in  the feature space


### Structure of CNN

#### Dense layer 
- classic fully connected neural network layer : each input node is connected to each output node
- learns global patterns in their input feature space

#### Convolution layer:
-  learns local patterns such as edges, textures,..
-  patterns that are learnt are translation invariant
	- once pattern is learnt,it can be recognized anywhere in the picture by the convnet
	- densely connected network would have to learn the pattern as the new one if it appeared at a new location
- learns spatial hierarchies of patterns
    - first learn small local patterns such as edges,the use them to discover the larger patterns made of the features of the first layers
- operates over 3D tensors - feature maps (height,width,depth)
    - in input layer depth means RGB value: red, green, and blue 

<img src="convNet0.png">

### How does CNN work:
1.  slide windows of size 3 × 3 or 5 × 5 over the 3D input feature map and stop at every possible location and extract the 3D patch
2.  transform 3D patch using filter into  into a 1D vector (tensor product with the same learned weight matrix, called the convolution kernel (filter))
3.  reassemble 1D vectors into a 3D output feature map of shape (height, width, output_depth) where output_depth-nr of filters


#### Size of the patches extracted from the inputs
-  typically 3 × 3 or 5 × 5

#### Depth of the output feature map:
-  number of filters computed by the convolution


<img src="dotProduct.png">
<img src="responseMap.png">

#### Remark
- output width and height may differ from the input width and height because of :
   - border effects	
   - use of strides

#### Border effect
- the weight and height of the out map decreases
- the problem can solved with the use of padding
<img src="borderEffect.png">

#### Padding
- adding an appropriate number of rows and columns on each side of the input feature map 
- Keras paremeters in Conv2D layer: "valid"= no padding,"same"= leave the orginal weight\height size 
<img src="padding.png">


#### Convolution strides
- distance between two successive windows 
- using stride 2 means the width and height of the feature map are downsampled by a factor of 2 + border effect 1 
- rarely used in practice

<img src="strides.png">


#### Max-Pooling operation:
- extracting windows from the input feature maps and outputting the max value
- downsample feature maps size (feature maps is halved after every MaxPooling2D layer) 
- max pooling is usually done with 2 × 2 windows and stride 2

<img src="maxPooling.png">


#### What happens if we remove max-pooling:
- features from the last convolution layer will not contain information about the totality of the input (can't recognize the digit basing on information from 7x7 pixel window)
- let's consider final feature map [22, 22, 64] which has 30,976 coefficients per sample. If you were to flatten it to stick a Dense layer of size 512 on top, that layer would have 15.8 million parameters!!!


#### Why do we apply downsampling:
- reduce the number of feature-map coefficients 
- induce spatial-filter hierarchies by making successive convolution layers look at increasingly large windows

#### Downsampling techniques:
- max pooling
- average pooling (average value of each channel over the patch)
- strides

#### Max-pooling vs Average-pooling:
- it’s more informative to look at the maximal presence of different features than at their average presence

- width_shift and height_shift: ranges (as a fraction of total width or height) within which to randomly translate pictures vertically or horizontally.
- shear_range: randomly applying shearing transformations.
- zoom_range : randomly zooming inside pictures.
- horizontal_flip: randomly flipping half the images horizontally—relevant when there are no assumptions of horizontal asymmetry (for example, real-world pictures).
- fill_mode: strategy used for filling in newly created pixels, which can appear after a rotation or a width/height shift

## Toy Example(MNIST digits)


#### MNIST Handwritten Digit Classification Dataset:
- Modified National Institute of Standards and Technology dataset
- 60,000 small square 28×28 pixel grayscale images of handwritten single digits
<img src="mnist.png">

#### Build  CNN model

```python
#Toy Example(MNIST digits):

from keras import layers
from keras import models

#define model
model = models.Sequential()
#define convolutional layer
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
#define convolutional layer
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
#define convolutional layer
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

#Add classfier at the top
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

#### Train model

```python
#Training the convnet on MNIST images
from keras.datasets import mnist
from keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype('float32') / 255

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=5, batch_size=64)
```

#### Evaluate model

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("accuracy:",test_acc)
```

## Deriving Convolution from a Fully Connected Layer

- Assuming that there are K neurons in the ﬁrst hidden layer, each neuron $H_i$ i , i = 0 . . . K in the hidden layer is
connected to all pixels in the grayscale image leading to W × H connections only for $H_i$

<img src="convNet.png">

- 16 × 16 grayscale image and 7200 neurons in first layers give : 7,372,800 distinct parameters !

#### How to reduce the number of paramaters ?

    1. The neuron in hidden layer arrange into 50 blocks of size 12 × 12

<img src="convNet2.png">

    2. Neurons in each block connect locally to the input image

<img src="convNet3.png">

    3. Neurons in one block can share the same set of weights

<img src="convNet4.png">

    4. Summary: The number of parameters is reduced to ( 5 × 5 ) × 50 × 12 × 12 = 180,000 which 97.5% reduction

Pixel ( m , n ) in an image is stronger correlated with its close neighbors than its far neighbors.
Neuron ( 0 , 0 ) in each block is intended to extract information from a region around pixel ( 2 , 2 ) in the image.


# Visualization of covnets learning process

### The representations learned by convnets can be visualized using one of the methods below:
- Visualizing intermediate convnet outputs: shows how successive convnet layers transform their input
- Visualizing convnets filters : shows precisely what visual pattern or concept each filter in a convnet is receptive to
- Visualizing heatmaps of class activation in an image: shows which parts of an image were identified as belonging to a given class

### Visualizing intermediate convnet outputs (INTERMEDIATE ACTIVATIONS)
- displays the feature maps that are output by various convolution and pools layers in a network  
- allows to understand how successive convnet layers transform their input
- visualizes the feature maps with three dimensions: width, height, and depth (channels)
- each channel encodes relatively independent feature

```python
#load image
img_path = 'cat2.png'
from keras.preprocessing import image
import numpy as np
img = image.load_img(img_path, target_size=(150, 150))
img_tensor = image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
img_tensor /= 255.

print("Tensor shape:",img_tensor.shape)

#Show image
import matplotlib.pyplot as plt
plt.imshow(img_tensor[0])
plt.show()
```

#### Visualize 4th and 7th channel

```python
from keras import models
layer_outputs = [layer.output for layer in model.layers[:8]]
activation_model = models.Model(inputs=model.input, outputs=layer_outputs)

activations = activation_model.predict(img_tensor)
first_layer_activation = activations[0]
```

```python
#Visualizing the fourth channel
import matplotlib.pyplot as plt
plt.matshow(first_layer_activation[0, :, :, 4], cmap='viridis')
print("\nIn my case the forth channel detects diagonal edges.\nBut your result can differ while specific filters learned by CNN are not deterministic.")
```

```python
# Seventh channel of the activation of the first layer on the test cat
plt.matshow(first_layer_activation[0, :, :, 7], cmap='viridis')
```

#### Visualize every channel

```python
#Visualizing every channel in every intermediate activation

layer_names = []                                                        
for layer in model.layers[:8]:                                          
    layer_names.append(layer.name)                              

images_per_row = 16

for layer_name, layer_activation in zip(layer_names, activations):
    n_features = layer_activation.shape[-1] 

    size = layer_activation.shape[1]    

    n_cols = n_features // images_per_row                           
    display_grid = np.zeros((size * n_cols, images_per_row * size))

    for col in range(n_cols):
        for row in range(images_per_row):
            channel_image = layer_activation[0,
                                             :, :,
                                             col * images_per_row + row]
            channel_image -= channel_image.mean()  
            channel_image /= channel_image.std()
            channel_image *= 64
            channel_image += 128
            channel_image = np.clip(channel_image, 0, 255).astype('uint8')
            display_grid[col * size : (col + 1) * size,
                         row * size : (row + 1) * size] = channel_image

    scale = 1. / size
    plt.figure(figsize=(scale * display_grid.shape[1],
                        scale * display_grid.shape[0]))
    plt.title(layer_name)
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')
```

#### Summary
- the first layer acts as a collection of various edge detectors. The activations retain almost all of the information present in the initial picture
- the higher activations layers become increasingly abstract and less visually interpretable
- in the following layers, more and more filters are blank
- blank square means the pattern encoded by the filter isn’t found in the input image
- specific filters learned by convolution layers aren’t deterministic

### Visualizing heatmaps of  class activation
- shows which parts of a given image led a convnet to its final classification decision
- class activation map (CAM) visualization technique:
  - 2D grid of scores associated with a specific output class
  - computed for every location in any input image
- for example: in cat classification problem: indicats how cat-like different parts of the image are

#### Grad-CAM: Visual Explanations from Deep Networks via Gradient-based Localization
-  weights every channel in feature map by the gradient of the class with respect to the channel
-  weights a spatial map of “how intensely the input image activates different channels” by “how important each channel is with regard to the detected class”

#### Load VGG16 model

```python
#load VGG16 model
from keras.applications.vgg16 import VGG16
model = VGG16(weights='imagenet')

from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
```

#### Load image and make prediction 

```python
#load image
img_path = 'elephant1.png'
img = image.load_img(img_path, target_size=(224, 224))

#createfloat32 Numpy array of shape (224, 224, 3)
x = image.img_to_array(img)

#Adds a dimension to transform the array into a batch of size
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

#make prediction for selected image
preds = model.predict(x)
print('most probably 3 classes:\n', decode_predictions(preds, top=3)[0])
# 386 is the number of the african elephant class
print('predicted class number:\n',np.argmax(preds[0]))
```

#### Create heatmap

```python
from keras import backend as K

# Get "African elephant” entry from the prediction vector
african_elephant_output = model.output[:, 386]

#Output feature map of the block5_conv3 layer,the last convolutional layer in VGG16
last_conv_layer = model.get_layer('block5_conv3')

#Gradient of the “African elephant” class with regard to the output feature map of block5_conv3
grads = K.gradients(african_elephant_output, last_conv_layer.output)[0] 

#Vector of shape (512,), where each entry is the mean intensity of the gradient over a specific feature-map
pooled_grads = K.mean(grads, axis=(0, 1, 2))
print("pooled_grads:\n",pooled_grads)

#returns function which evaluates outputs([pooled_grads, last_conv_layer.output[0]]) using inputs
iterate = K.function([model.input],[pooled_grads, last_conv_layer.output[0]])
#Get two Numpy arrays for image X
pooled_grads_value, conv_layer_output_value = iterate([x])

#Multiply each channel in the feature-map array by importance of the channel
for i in range(512):
    conv_layer_output_value[:, :, i] *= pooled_grads_value[i]

#The channel-wise mean of the resulting feature map is the heatmap of the class activation.
heatmap = np.mean(conv_layer_output_value, axis=-1)                    
```

#### Normalize heatmap

```python
#Normalize values between 0 and 1
heatmap = np.maximum(heatmap, 0)
heatmap /= np.max(heatmap)
#Show heatmap
plt.matshow(heatmap)
```

#### Join the heatmap and original  image

```python
#install cv2 (opencv) is neccessary
#!pip install opencv-python
import cv2
import matplotlib.image as mpimg
img_name='elephant_heatmap.png'

img = cv2.imread(img_path)
heatmap2 = cv2.resize(heatmap, (img.shape[1], img.shape[0]))

#Converts the heatmap to RGB
heatmap2 = np.uint8(255 * heatmap2)
#Create colorMap using heatmap,higher value mean more reddishcolor
heatmap2 = cv2.applyColorMap(heatmap2, cv2.COLORMAP_JET)

#join heatmap and image
superimposed_img = heatmap2 * 0.4 + img
cv2.imwrite(img_name, superimposed_img)

#show image
img= mpimg.imread(img_name)
plt.imshow(img)
print("\nThe most informative area of the image is the head of small elephant.")
```

#### Summary
- heatmap methods explains why African elephant was found in the image
- finds the location of African elephant on the image

# Text vectorization

#### Text vectorization
- process of transforming text into numeric tensors

#### Vectorization methods
- segment text into words, and transform each word into a vector
- segment text into characters, and transform each character into a vector
- extract n-grams of words or characters, and transform each n-gram into a vector


<img src="tokenization.png">

####  N-grams
- overlapping groups of multiple consecutive words or characters that you can extract from a sentence
- popular form of feature engineering
- used in shallow language-processing models

tokens =   outputs of breaking down text into words, characters, or n-grams <br>
tokenization = process of breaking text into tokens <br>


#### Example Word N-grams
- input= The cat sat on the mat
- ouptut2 {"The", "The cat", "cat", "cat sat", "sat",
  "sat on", "on", "on the", "the", "the mat", "mat"}
- output3={"The", "The cat", "cat", "cat sat", "The cat sat",
  "sat", "sat on", "on", "cat sat on", "on the", "the",
  "sat on the", "the mat", "mat", "on the mat"}

#### N-gram in Deep Learning?
- n-grams are not used in deep learning, while CNN and RNN allow for hierarchical feature learning
- n-grams are popular in shallow learning (decision trees, SVM, Naive Bayes,...)

### How  associate a vector with a token?
- one-hot encoding 
- token embedding

#### One-Hot encoding:
- associate a unique integer index with every word 
- turn integer index i into a binary vector of size N (the size of the vocabulary)
- done also on character level 
- possible extension: one-hot hashing trick
- BUT: we encounter two problems: many zeros +  high dimensionality of the vector

#### Word Embeddings:
- known also as dense word vectors
- learned from data
- overcome the problems of one-hot encoding
    - binary, sparse (mostly made of zeros)
    - very high-dimensional:  256-dimensional vs 20,000-dimensional
    
<img src="embedding0.png">  
    
    
##### How to create word embeddings?
- naive approach: use random vector for given word
     - problem with synonyms 
- remember that semantic relationships varies from task to task 
- learn a new embedding space with every new task

<img src="embedding1.png">

##### Embedding Layer 
- dictionary that maps integer indices (which stand for specific words) to dense vectors

##### Algorithm:
1. input =2D tensor of integers, of shape (samples, sequence_length)
2. each sample is a sequence of integers (32, 10) = batch of 32 sequences of length 10 
3. returns a 3D floating-point tensor of shape (samples, sequence_length, embedding_dimensionality)
4. 3D tensor is processed by an RNN layer or a 1D convolution layer 



#### Pretrained word embeddings
- use when the input data set is small
- precomputed databases of word embeddings
    - Word2vec
        - two-layer neural net that processes text
        - its input is a text corpus and its output is a set of vectors: feature vectors for words in that corpus 
    - Global Vectors for Word Representation (GloVe, https://nlp.stanford.edu/projects/glove)    
    


```python
#Embedding layer takes at least two arguments: the number of possible tokens 
#(here, 1,000: 1 + maximum word index) and the dimensionality of the embeddings (here, 64).
from keras.layers import Embedding

embedding_layer = Embedding(1000, 64) 
```

# Sequence processing with covnets

- highly relevant to sequence processing
- time can be treated as a spatial dimension
- 1D convnets can be competitive with RNNs (cheaper computational cost)
- we can use pooling techniques:max-pooling,average-pooling

#### Remarks:
- 1D convnet processing sequences of characters size 5 should be able to learn words or word fragments of length 5 or less
- 1D convnet is able to recognize these words in any context in an input sequence

### Applications of 1D covnets
- audio generation 
- machine translation

#### Algorithm
1. extract 1D patches from timseries tensors
2. apply an identical transformation to every patch

### CNNs and RNNs to process long SEQUENCES

### Drawback's of CNNs
- 1D covnets process input patches independently, they aren’t sensitive to the order unlike RNNs
- more recent data points are not interpreted differently from older data points 
- applying many convolution layers and pooling layers does not solve the problem

#### Solution
- combine CNNs and RNNs. Useful approach for long timerseries, which can't be proccededd with RNNs 


<img src="cnnRnn.png">

#### Advantages
- faster then RNNs based approach
- able to use more input data

#### Disadvantages
